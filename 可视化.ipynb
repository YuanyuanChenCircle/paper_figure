```python
accu = [0.4, 0.5, 0.6, 0.7]
temp= accu

plt.figure(figsize=(5.9, 5.3))

tick = ['f', 'd', 'g', 'h']
bar_width = 0.3
font2 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 12,
}
font3 = {'family' : 'Times New Roman',
'weight' : 'bold',
'size'   : 12,
}

plt.bar(range(len(tick)),temp,tick_label = tick, width = bar_width)
plt.xlabel('haha',font2)
plt.ylabel('sss',font2)
plt.tick_params(labelsize=12) #刻度字体大小13
for x,y in enumerate(temp):
    plt.text(x-0.14,y+0.010,'%s' %round(y,3), va='center', font=font3)

plt.savefig('./1.svg',dpi=600,bbox_inches = 'tight',pad_inches = 0)

plt.show()

```
